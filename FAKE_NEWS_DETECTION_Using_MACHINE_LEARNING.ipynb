```python
import re
```

```python
def wordopt(text):
  # Convert into lowercase
  text= text.lower()

  # Remove URLs
  text = re.sub(r"https?://\S+|www\.\S","",text)

  # Remove HTML tags
  text =re.sub(r"[^\w\s]","",text)

  # Remove digits
  text = re.sub(r"\d","",text)

  # Remove newline characters
  text = re.sub(r"\n"," ",text)

  return text
```

```python
def output_lable(n):
  if n == 0:
    return "Fake News"
  elif n == 1:
    return "True News"
```
