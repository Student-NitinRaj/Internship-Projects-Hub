```python
import pandas as pd

# Assuming your DataFrame is named 'df'
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DataSet/Titanic Data.xlsx")

# Basic Descriptive Statistics
print("Descriptive Statistics:\n", df.describe())

# Count of Missing Values per Column
print("\nMissing Values per Column:\n", df.isnull().sum())

# Most Frequent Values
print("\nMost Frequent Values:\n", df.mode().iloc[0])  # .iloc[0] gets the first row (mode)

# Correlation between 'balance' and 'income' (if applicable)
if 'balance' in df.columns and 'income' in df.columns:
    correlation = df['balance'].corr(df['income'])
    print("\nCorrelation between Balance and Income:", correlation)

# Summary of Data Types
print("\nData Types:\n", df.dtypes)
```
